## paragraph to sentence tokenization

```python
import nltk
```

```python
sentence = nltk.sent_tokenize(paragraph)
```

```python
sentence
```

## word tokenization

```python
words = nltk.word_tokenize(paragraph)
```

```python
words
```

## stemming

```python
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
import re
```

```python
stemmer = PorterStemmer()
```

```python
for i in range(len(sentence)):
    review = re.sub('[^a-zA-Z]', ' ', sentence[i])
    review = review.lower()
    review = review.split()
#     print(review)
    words = [stemmer.stem(word) for word in review if word not in stopwords.words("english")] ## converted in a list
    print(words)
```

## Lemmatization

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
for i in range(len(sentence)):
    review = re.sub('[^a-zA-Z]', ' ', sentence[i])
    review = review.lower()
    review = review.split()
#     print(review)
    words = [lemmatizer.lemmatize(word) for word in review if word not in stopwords.words("english")] ## converted in a list
    print(words)
```

## word to vec

```python
from gensim.models import Word2Vec
```

```python
# Preparing the dataset
sentences = nltk.sent_tokenize(paragraph)

sentences = [nltk.word_tokenize(sentence) for sentence in sentences]
# print(sentences)
for i in range(len(sentences)):
    review = re.sub('[^a-zA-Z]', ' ', sentence[i])
    review = review.lower()
    review = review.split()
    sentences[i] = [word for word in review if word not in stopwords.words('english')]
```

```python
model = Word2Vec(sentences, min_count=1)
```

```python
model.wv.most_similar("ai")
```
